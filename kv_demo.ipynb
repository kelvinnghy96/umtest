```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Load your dataset
data = pd.read_csv(".\\Global Superstore.csv")  # Update with your dataset filename

# Detect numeric columns
numeric_columns = data.select_dtypes(include=['int', 'float']).columns

# Calculate number of rows and columns for the grid
num_plots = len(numeric_columns)
num_rows = math.ceil(num_plots / 3)
num_cols = min(num_plots, 3)

# Create a grid of subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))

# Flatten the axes array if it's more than 1 dimension
if num_plots > 1:
    axes = axes.flatten()

# Plot histograms for each numeric column
for i, column in enumerate(numeric_columns):
    ax = axes[i] if num_plots > 1 else axes
    ax.hist(data[column], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(column)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Remove any extra empty subplots
for i in range(num_plots, num_rows * num_cols):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Function to check if a number is positive, negative, or zero
def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

# Input the first number
num1 = float(input("Enter the first number: "))

# Input the second number
num2 = float(input("Enter the second number: "))

# Perform addition
addition_result = num1 + num2
print("Addition result:", addition_result, "(", check_number(addition_result), ")")

# Perform subtraction
subtraction_result = num1 - num2
print("Subtraction result:", subtraction_result, "(", check_number(subtraction_result), ")")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pdfkit

# Load your dataset
data = pd.read_csv("Global Superstore.csv")

# Perform Exploratory Data Analysis (EDA)
# Univariate Analysis
univariate_analysis = data.describe(include='all')

# Pairplot
pairplot = sns.pairplot(data)
pairplot.savefig("pairplot.png")

# Save EDA and pairplot results to PDF
pdf_filename = "EDA_Report.pdf"
html_template = """
<!DOCTYPE html>
<html>
<head>
    <title>Exploratory Data Analysis Report</title>
    <style>
        h1 {{
            text-align: center;
        }}
        h2 {{
            margin-top: 30px;
        }}
        p {{
            margin-top: 10px;
        }}
        img {{
            display: block;
            margin-left: auto;
            margin-right: auto;
            width: 50%;
        }}
    </style>
</head>
<body>
    <h1>Exploratory Data Analysis Report</h1>
    <img src="logo.png" alt="Your Logo">
    
    <h2>Univariate Analysis</h2>
    {}
    
    <h2>Pairplot</h2>
    <img src="pairplot.png" alt="Pairplot">
</body>
</html>
"""

with open("univariate_analysis.txt", "w") as text_file:
    text_file.write(univariate_analysis.to_string())

html_content = html_template.format(univariate_analysis.to_html())

with open("report.html", "w") as html_file:
    html_file.write(html_content)

# Convert HTML to PDF
pdfkit.from_file("report.html", pdf_filename)

# Clean up intermediate files
os.remove("report.html")
os.remove("univariate_analysis.txt")

print("PDF report generated successfully.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from reportlab.lib import colors
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet

# Load your dataset
data = pd.read_csv("Global Superstore.csv")

# Perform Exploratory Data Analysis (EDA)
# Univariate Analysis
univariate_analysis = data.describe(include='all')

# Pairplot
pairplot = sns.pairplot(data)
pairplot.savefig("pairplot.png")

# Save EDA and pairplot results to PDF
pdf_filename = "EDA_Report.pdf"
logo_path = "logo.png"

# Create a PDF document
doc = SimpleDocTemplate(pdf_filename, pagesize=letter)
styles = getSampleStyleSheet()
style_heading = styles["Heading1"]
style_normal = styles["Normal"]

# Add logo and title
title = "Exploratory Data Analysis Report"
logo = Image(logo_path, width=150, height=150)
logo.hAlign = 'CENTER'
doc_title = Paragraph(title, style_heading)
doc.build([logo, doc_title])

# Add univariate analysis
univariate_data = [Paragraph(text, style_normal) for text in univariate_analysis.to_string().split("\n")]
doc.build([Paragraph("Univariate Analysis", style_heading), Spacer(1, 12)] + univariate_data)

# Add pairplot
pairplot_img = Image("pairplot.png", width=500, height=500)
pairplot_img.hAlign = 'CENTER'
doc.build([Paragraph("Pairplot", style_heading), Spacer(1, 12), pairplot_img])

print("PDF report generated successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

def univariate_analysis(df):
    num_cols = df.shape[1]
    num_rows = math.ceil(num_cols / 3)
    fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows*5))
    axes = axes.flatten()

    for i, col in enumerate(df.columns):
        ax = axes[i]
        if df[col].dtype == 'object':  # categorical column
            if df[col].nunique() <= 10:
                sns.countplot(data=df, x=col, ax=ax, palette='Set3')
            else:
                top_values = df[col].value_counts().nlargest(10)
                sns.countplot(data=df[df[col].isin(top_values.index)], x=col, ax=ax, palette='Set3')
        else:  # numerical column
            df[col].plot(kind='hist', ax=ax, title=col)
        ax.set_ylabel('Frequency')
        ax.set_xlabel(col)

    # Remove any unused subplots
    for i in range(num_cols, num_rows * 3):
        fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()

# Assuming your dataframe is named 'df'
univariate_analysis(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

def univariate_analysis(df):
    num_cols = df.shape[1]
    num_rows = math.ceil(num_cols / 3)
    fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows*5))
    axes = axes.flatten()

    # Define a color palette
    colors = sns.color_palette("husl", 10)

    for i, col in enumerate(df.columns):
        ax = axes[i]
        if df[col].dtype == 'object':  # categorical column
            if df[col].nunique() <= 10:
                df[col].value_counts().plot(kind='bar', ax=ax, title=col, color=colors)
            else:
                top_values = df[col].value_counts().nlargest(10)
                top_values.plot(kind='bar', ax=ax, title=col, color=colors)
        else:  # numerical column
            df[col].plot(kind='hist', ax=ax, title=col)
        ax.set_ylabel('Frequency')
        ax.set_xlabel(col)

    # Remove any unused subplots
    for i in range(num_cols, num_rows * 3):
        fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()

# Assuming your dataframe is named 'df'
univariate_analysis(df)
```

```python
import os

def count_files_in_folders(directory):
    folder_details = {}

    for root, dirs, files in os.walk(directory):
        relative_path = os.path.relpath(root, directory)
        folder_details[relative_path] = len(files)

    return folder_details

def compare_folder_details(details1, details2):
    all_folders = set(details1.keys()).union(set(details2.keys()))
    differing_folders = {}

    for folder in all_folders:
        count1 = details1.get(folder, 0)
        count2 = details2.get(folder, 0)
        if count1 != count2:
            differing_folders[folder] = (count1, count2)

    return differing_folders

def print_differing_folders(differing_folders):
    for folder, (count1, count2) in differing_folders.items():
        print(f"{folder}: Folder 1 has {count1} files, Folder 2 has {count2} files")

# Replace these paths with your OneDrive folder paths
one_drive_folder_1 = r"C:\Users\kelvi\OneDrive\0 Kelvin's PC"
one_drive_folder_2 = r"C:\Users\kelvi\OneDrive - Asia Pacific University\0 Kelvin's PC"

folder_details_1 = count_files_in_folders(one_drive_folder_1)
folder_details_2 = count_files_in_folders(one_drive_folder_2)

differing_folders = compare_folder_details(folder_details_1, folder_details_2)

print("Folders with differing file counts:")
print_differing_folders(differing_folders)
```
