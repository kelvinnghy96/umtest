```python
import matplotlib.pyplot as plt
import numpy as np

category = ['Market', 'Category']
numeric = ['Profit', 'Sales']

fig, axs = plt.subplots(2, 2, figsize=(10, 10))  # Creating a 2x2 grid of subplots

# Creating bar charts
for i, y in enumerate(category):
    for j, x in enumerate(numeric):
        bars = axs[i, j].bar(df5[y], df5[x])
        col = y + ', ' + x
        axs[i, j].set_title(col)  # Adding title based on the current value of x and y

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your DataFrame containing 'Market' and 'Profit' columns

# Extracting unique markets and their corresponding profits
unique_markets = df['Market'].unique()
profits_by_market = [df[df['Market'] == market]['Profit'].sum() for market in unique_markets]

# Generating distinct colors for each market
colors = plt.cm.tab10(np.arange(len(unique_markets)))

# Creating bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(unique_markets, profits_by_market, color=colors)

# Adding labels and title
plt.xlabel('Market')
plt.ylabel('Profit')
plt.title('Profit by Market')

# Add data labels above each bar
for bar, profit in zip(bars, profits_by_market):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05, f'{profit:.2f}', ha='center', va='bottom')

# Customizing legend
plt.legend(bars, unique_markets)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```
